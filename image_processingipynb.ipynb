```python
#resizing the image using pillow library
from PIL import Image
```

```python
img = Image.open('/content/puppy.jpg')
img_resized=img.resize((200,200))
img_resized.save('puppy1.jpg')
```

```python
# importing OpenCV library
import cv2
```

```python
img = cv2.imread('/content/puppy1.jpg')
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
from google.colab.patches import cv2_imshow
```

```python
# displaying the image
cv2_imshow(grayscale_image)
```

```python
# saving the grayscale image
cv2.imwrite('puppy1_grayscale.jpg', grayscale_image)
```
